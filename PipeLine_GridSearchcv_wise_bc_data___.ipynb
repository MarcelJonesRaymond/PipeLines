```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.model_selection import  GridSearchCV,train_test_split,StratifiedKFold,cross_val_score
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.svm import SVC,SVR
from sklearn import metrics
from sklearn.decomposition import PCA
```

```python
data= pd.read_csv('wisc_bc_data.csv')
```

```python
label=LabelEncoder()
data['diagnosis']=label.fit_transform(data['diagnosis'])
```

```python
x=data.drop('diagnosis',1)
y=data[['diagnosis']]
```

```python
X_train,x_test,Y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
svm=SVC()
result=cross_val_score(svm,X_train,Y_train,cv=10)
```

```python
result.mean()
```

```python
result.std()
```

```python
kfold=StratifiedKFold(n_splits=10)
for train,test in kfold.split(x,y):
    X_train_sm,Y_train_sm=x.loc[train],y.loc[train]
    x_test_sm,y_test_sm=x.loc[test],y.loc[test]
```

```python
import warnings
warnings.filterwarnings('ignore')
cv=StratifiedKFold(n_splits=10)
result=cross_val_score(svm,X_train_sm,Y_train_sm,cv=cv)
print('Model performance  with the avg accuracy of',result.mean(), '  and STD of ',result.std())
```

```python
pipe=make_pipeline(StandardScaler(),PCA(2),SVC())
```

```python
pipe.fit(X_train_sm,Y_train_sm)
yhat_pipe=pipe.predict(x_test_sm)
print(metrics.classification_report(y_test_sm,yhat_pipe))
```

```python
result=cross_val_score(pipe,X_train,Y_train,cv=10)
print('Model performance  with the avg accuracy of',result.mean(), '  and STD of ',result.std())
```

```python
param_grid = {'pca__n_components':[14,15],'svc__C': [0.001, 0.01, 0.1, 1, 10, 100], 'svc__gamma': [0.001, 0.01, 0.1, 1, 10], 'svc__kernel':['rbf','poly']} 
```

```python
model=GridSearchCV(pipe,param_grid,cv=10)
model.fit(X_train_sm,Y_train_sm)
model.best_params_
```

```python
yhat_best=model.predict(x_test_sm)
print(metrics.classification_report(y_test_sm,yhat_best))
```

```python
print(metrics.confusion_matrix(y_test_sm,yhat_best))
```
